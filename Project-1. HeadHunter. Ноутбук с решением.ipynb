3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def experience_to_months(s):
    """ Функция преобразует строковую запись из столбца 'Опыт работы' в количество месяцев опыта
    Аргумент: s - строковая запись из столбца 'Опыт работы'
    Возвращаемое значение: число в формате int, равное количеству месяцев опыта
    """
    if pd.isna(s):
        return pd.NA # NaN оставляем как NaN
    elif s == 'Не указано':
        return pd.NA  # пропуски заполняем NaN
    else:
        months = 0 # количество месяцев, указанное в записи  
        years = 0 # количество лет, указанное в записи
        
        l = s.split()[2:6] #разделяем строку на список слов, оставляем с 3 по 6 слова - в них информация о сроке
        if l[3][:5] == 'месяц': # если последнее слово начинается с 'месяц' - значит указано и количество лет и количество месяцев
            months = int(l[2])
            years = int(l[0])
        elif l[1][:5] == 'месяц': # иначе - занчит указаны либо месяцы, либо годы. Проверяем 2 слово, если начинается с 'месяц' - значит указано только количество месяцев
            months = int(l[0])
        else: # значит указано только количество лет
            years = int(l[0])
        return int(years * 12 + months)

hh_raw_data['Опыт работы (месяц)'] = hh_raw_data['Опыт работы'].apply(experience_to_months) 

hh_raw_data.drop('Опыт работы', axis=1, inplace=True); #удаляем признак 'Опыт работы'

hh_raw_data['Опыт работы (месяц)'].median() #определяем медианный опыт работы в месяцах
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#города-миллионники
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

def split_and_extract_geography(s):
    """ Функция преобразует строковую запись из столбца 'Город, переезд, командировки' в 
    список значений для столбцов 'Город', 'Готовность к переезду', 'Готовность к командировкам'    
    
    Аргумент: s - строковая запись из столбца 'Город, переезд, командировки', содержит запись в формате 
    <Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>
    Необязательные компоненты в скобках.
    
    Возвращаемое значение: список из трех компонент:
    'Город' - название города если это Москва или Санкт-Петербург, "город-миллионник" если город из списка, иначе - "другие" 
    'Готовность к переезду' - True если готов к переезду, False - иначе, 
    'Готовность к командировкам' - True если готов к командировкам, False - иначе
    """
    city = '' # город после преобразования
    movint_to_another_city = False # готовность к переезду
    ready_for_business_trips = False # готовность к командировкам, по умолчанию, если не найдено - не готов
    #display(s)
    l = s.split(' , ') # по этому разделителю строка с данными распадается на до 4 фрагментов: 
    # в 1ом всегда указан город, в ппоследних двух точно есть данные про переезд и возможно про командировки (если они заполнены и не потеряны)
    if l[0] == 'Москва' or l[0] == 'Санкт-Петербург': # проверяем и обрабатываем город
        city = l[0]
    elif l[0] in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
    for i in l[-2:]: # проверяем и обрабатываем информацию про переезд и командировку
        if i.find('перее') >= 0 and i.find('не готов') == -1: # фрагмент про переезд, и если не содержит "не готов" - меняем на True, по умолчанию остается False
            movint_to_another_city = True
        elif i.find('командиров') >= 0 and i.find('не готов') == -1: # фрагмент про командировку, и если не содержит "не готов" - меняем на True, по умолчанию остается False
            ready_for_business_trips = True
    return [city, movint_to_another_city, ready_for_business_trips]
    

temp = hh_raw_data['Город, переезд, командировки'].apply(split_and_extract_geography) # применяем функцию, получаем столбец Series со списками готовых значений
# далее преобразуем Series в список, и трансформируем стандартным инициализатором DatarFrame в 3 столбца
hh_raw_data[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = pd.DataFrame(list(temp)) #преобразуем Series в список, получаем список строк, создаем изнего Dataframe

hh_raw_data.drop('Город, переезд, командировки', axis=1, inplace=True); #удаляем признак 'Город, переезд, командировки'

# отвечаем на вопросы задачи
display(hh_raw_data['Город'].value_counts(normalize=True) * 100)
display(hh_raw_data[['Готовность к переезду', 'Готовность к командировкам']].value_counts(normalize=True) * 100)

```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#создаем списки с названиями столбцов признаков-мигалок
job_timing = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка'] #признаки занятости
schedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод'] #признаки графика

for i in job_timing: # ставим True в столбец с соответствующим типом занятости только если этот тип есть в столбце "Занятость"
    hh_raw_data[i] = hh_raw_data['Занятость'].apply(lambda x: i in x)  

for i in schedule:  # ставим True в столбец с соответствующим типом графика только если этот тип есть в столбце "График"
    hh_raw_data[i] = hh_raw_data['График'].apply(lambda x: i in x)

hh_raw_data.drop('Занятость', axis=1, inplace=True); #удаляем признак 'Занятость'
hh_raw_data.drop('График', axis=1, inplace=True); #удаляем признак 'График'

# отвечаем на вопросы задачи
display(hh_raw_data[['проектная работа', 'волонтерство']].value_counts())
display(hh_raw_data[['вахтовый метод', 'гибкий график']].value_counts())
```

* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.
Воспользуйтесь этой информацией в ваших преобразованиях.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#читаем историю курсов валют из файла
currency_data = pd.read_csv('./data/ExchangeRates.csv')

hh_raw_data[['ЗП сумма', 'currency']] = hh_raw_data['ЗП'].str.split(expand=True) # разделяем сумму зарплаты и ее валюту по пробелу и записываем в соответствующие столбцы

# словарь-справочник для приведения валют из базы HH.ru к кодам - формату, используемому в файле с историей курсов валют;
# единственная валюта, которой нет в файле с историей курсов - рубли
curr_codes = {
    'руб.'    : 'RUB',
    'KZT'     : 'KZT',
    'USD'     : 'USD',
    'бел.руб.': 'BYN',
    'EUR'     : 'EUR',
    'грн.'    : 'UAH',
    'сум'     : 'UZS',
    'KGS'     : 'KGS',
    'AZN'     : 'AZN'
}
currency_data['date'] = pd.to_datetime(currency_data['date']).dt.date # приводим столбец к типу данных datetime
hh_raw_data['date'] = pd.to_datetime(hh_raw_data['Обновление резюме']).dt.date # приводим столбец к типу данных datetime
hh_raw_data['currency'] = hh_raw_data['currency'].map(curr_codes) # приводим валюты к кодам валют по словарю-справочнику
hh_raw_data['ЗП сумма'] = pd.to_numeric(hh_raw_data['ЗП сумма']) # приводим строковые значения сумм ЗП к числовому формату

# используем левое присоединение: для каждой строки из базы HH.ru по ключу валюты и даты присоединяем строку из файла с историей курсов валют 
hh_raw_data = hh_raw_data.merge( 
    right=currency_data[['currency', 'date', 'close', 'proportion']],
    on=['currency', 'date'],
    how='left',
    #suffixes=('None','_curr')
)

# пересчитываем в рубли суммы из других валют по формуле "сумма в валюте" * курс / пропорция; суммы в рублях - оставляем как есть (они не требуют конвертации)
hh_raw_data['ЗП (руб)'] = hh_raw_data.apply(lambda x: x['ЗП сумма'] * x['close'] / x['proportion'] if x['currency'] != 'RUB' else x['ЗП сумма']*1.00, axis=1)

# отвечаем на вопросы задачи
display(hh_raw_data['ЗП (руб)'].median())

hh_raw_data.drop(['ЗП', 'ЗП сумма', 'currency', 'date', 'close', 'proportion'], axis=1, inplace=True); #удаляем столбец 'ЗП' и все промежуточные
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
#создаем переменную-столбец слогарифмом возраста
log_age = np.log(hh_raw_data['Возраст'])

fig, axes = plt.subplots(figsize=(8, 4))
axes.set_title('Логарифм возраста соискателей')

#гистограмма с 40 "корзинами" по логарифму возраста
axes.minorticks_on() 
axes.xaxis.set_tick_params(rotation=30)
axes.grid(which='both', axis='x', dashes=[6,2])
sns.histplot(
    data=log_age,
    bins=20,
);

#вычисляем среднее, стандартное отклонение и левую/правую границу z-метода 
mu = log_age.mean()
sigma = log_age.std()
lower_bound = mu - 3 * sigma
upper_bound = mu + 4 * sigma

#выводим на график среднее распределения и границы z-метода 
axes.axvline(mu, color='b', linestyle='--', lw = 2)
axes.axvline(lower_bound, color='k', lw = 2)
axes.axvline(upper_bound, color='k', lw = 2)

#создаем маску для фильтрации выбросов
outliers = (log_age < lower_bound) | (log_age > upper_bound)

#выводим количество выбросок и асимметричность логарифмического распределения
display(outliers.sum())
display(log_age.skew())

#выводим таблицу с выбросами
display(hh_raw_data[outliers])

#очищаем данные от выбросов
hh_raw_data = hh_raw_data[~outliers]
```

Вывод: асимметрия логариaмического распределения правосторонняя, т.к. skew > 0. С учетом послабления в выбросы попали возрасты 15 и 100 лет.
